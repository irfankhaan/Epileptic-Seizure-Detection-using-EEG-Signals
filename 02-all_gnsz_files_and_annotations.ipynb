### Saving file path, onset and end time to a CSV file of the 256 Hz gnsz files

```python
import os
import numpy as np
import mne
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib qt
```

```python
fileObject = open('gnsz_256.txt', "r")
data = fileObject.read()
```

```python
gnsz_256 = data.split('\n')
```

```python
gnsz_256
gnsz_256.pop()
```

```python
pref_ch = ['EEG FP1-REF',
 'EEG FP2-REF',
 'EEG F3-REF',
 'EEG F4-REF',
 'EEG C3-REF',
 'EEG C4-REF',
 'EEG P3-REF',
 'EEG P4-REF',
 'EEG O1-REF',
 'EEG O2-REF',
 'EEG F7-REF',
 'EEG F8-REF',
 'EEG T3-REF',
 'EEG T4-REF',
 'EEG T5-REF',
 'EEG T6-REF',
 'EEG FZ-REF',
 'EEG CZ-REF',
 'EEG PZ-REF',
 'EEG A1-REF',
 'EEG A2-REF']
```

```python
gnsz_eeg_annot = dict()
```

```python
gnsz_256[0].split('/')[9]
```

```python
iter = 1
for path in gnsz_256:
    temp = dict()
    temp['path'] = path
    raw = mne.io.read_raw_edf(path[:-4] + '.edf')
    temp['edf'] = raw.pick_channels(pref_ch)

    fileObject = open(path[:-4] + '.tse', "r")
    data = fileObject.read()
    
    components = data.split('\n\n')
    events = components[1].split('\n')

    onset = []
    duration = []
    description = []
    end = []
    for i in range(len(events)-1):
        onset.append(float(events[i].split()[0]))
        end.append(float(events[i].split()[1]))
        duration.append(round(float(events[i].split()[1]) - float(events[i].split()[0]),4))
        description.append(events[i].split()[2])
    
    temp['onset'] = onset
    temp['end'] = end
    temp['duration'] = duration
    temp['description'] = description

    gnsz_eeg_annot[iter] = temp
    iter = iter+1
```

```python
gnsz_eeg_annot
```

```python
eg = gnsz_eeg_annot[44]
annotations = mne.Annotations(eg['onset'],eg['duration'],eg['description'])
raw_temp = eg['edf'].copy()
raw_temp.set_annotations(annotations)
raw_temp.plot()
eg['path']
```

```python
import pandas as pd
gnsz_data = pd.DataFrame.from_dict(gnsz_eeg_annot)
gnsz_data.to_csv("gnsz_eeg.csv")
```
