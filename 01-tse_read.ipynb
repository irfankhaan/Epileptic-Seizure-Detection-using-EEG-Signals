### Making a list of the files with only generalised seizure (gnsz) and sampling frequency of 256 Hz 

```python
import os
import os
import numpy as np
import mne
import matplotlib.pyplot as plt
%matplotlib qt
```

```python
root = "D:/TUH EEG Corpus/tuh_eeg/edf/train/01_tcp_ar"
```

```python
gnsz = []
fnsz = []
spsz = []
cpsz = []
absz = []
tnsz = []
tcsz = []
# r=root, d=directories, f = files
for r, d, f in os.walk(root):
    for file in f:
        if file.endswith(('tse')):
            fileObject = open(r+'/'+file, "r")
            data = fileObject.read()
            for i in data.split():
                if i == 'gnsz':
                    gnsz.append(r+'/'+file)
                if i == 'fnsz':
                    fnsz.append(r+'/'+file)
                if i == 'spsz':
                    spsz.append(r+'/'+file)
                if i == 'cpsz':
                    cpsz.append(r+'/'+file)
                if i == 'absz':
                    absz.append(r+'/'+file)
                if i == 'tnsz':
                    tnsz.append(r+'/'+file)
                if i == 'tcsz':
                    tcsz.append(r+'/'+file)
#files.sort(key=lambda f: int(''.join(filter(str.isdigit, f))))
files_gnsz = list(dict.fromkeys(gnsz))
files_fnsz = list(dict.fromkeys(fnsz))
files_spsz = list(dict.fromkeys(spsz))
files_cpsz = list(dict.fromkeys(cpsz))
files_absz = list(dict.fromkeys(absz))
files_tnsz = list(dict.fromkeys(tnsz))
files_tcsz = list(dict.fromkeys(tcsz))
```

```python
for i in files_gnsz:
    for j in files_fnsz:
        if i == j:
            print(i)
```

```python
from collections import Counter
res = list((Counter(files_gnsz)-Counter(files_fnsz)).elements())
res = list((Counter(res)-Counter(files_spsz)).elements())
res = list((Counter(res)-Counter(files_cpsz)).elements())
res = list((Counter(res)-Counter(files_absz)).elements())
res = list((Counter(res)-Counter(files_tnsz)).elements())
res = list((Counter(res)-Counter(files_tcsz)).elements())
```

```python
gnsz_file_path = []
for i in res:
    a = i.split()[2].split('\\')
    s_path = root +'/'+  a[1] +'/'+  a[2] +'/'+ a[3]
    gnsz_file_path.append(s_path)
```

```python
all_freq = []
gnsz_256 = []
for path in gnsz_file_path:
    raw = mne.io.read_raw_edf(path[:-4]+'.edf')
    all_freq.append(raw.info['sfreq'])
    if raw.info['sfreq'] == 256.0:
        gnsz_256.append(path)
```
