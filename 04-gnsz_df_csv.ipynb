## Filtering, windowing and extracting features from the gnsz segments

```python
import os
import numpy as np
import pandas as pd
import mne
import mne_features
import matplotlib.pyplot as plt
%matplotlib qt
```

# Importing EDF/FIF Files

```python
root = "../Dataset/gnsz/"
```

```python
for r, d, f in os.walk(root):
    for file in f:
        print(file)
```

```python
df = pd.DataFrame()
for file in f:
    temp_df = pd.DataFrame()
    raw = mne.io.read_raw_fif(root + file, preload=True)

    raw.notch_filter(freqs=np.arange(60,128,60))
    raw.filter(l_freq=0.5, h_freq=50, fir_window="blackman")

    epochs = mne.make_fixed_length_epochs(raw, duration=1)

    epoch_data = epochs.get_data()
    temp_df = mne_features.feature_extraction.extract_features(epoch_data, sfreq=raw.info['sfreq'], selected_funcs=['mean','std','variance','ptp_amp','skewness','kurtosis','hurst_exp','app_entropy','samp_entropy','hjorth_mobility','hjorth_complexity','line_length','svd_fisher_info','katz_fd','higuchi_fd','pow_freq_bands','hjorth_mobility_spect','hjorth_complexity_spect','spect_entropy','svd_entropy','energy_freq_bands','spect_edge_freq'],return_as_df=True)

    df = df.append(temp_df, ignore_index=True)
```

```python
df = df.assign(output=1)
```

```python
df.to_csv('../Dataset/gnsz_features_data.csv', index = False)
```
